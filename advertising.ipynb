# PreProccessing(using TF-idf for column)

### Omitting Extra column

```python
data = data.drop(columns=['date', 'time', 'Timestamp'])
data.head()
```

### Drop duplicate rows

```python
data = data.drop_duplicates()
data.shape
```

### Encoding Categorical data

```python
print(len(data['Ad Topic Line'].unique()))
print(len(data['City'].unique()))
print(len(data['Country'].unique()))
```

```python
vectorized_data = pd.get_dummies(data, columns=['City', 'Gender', 'Country'])

for col in vectorized_data.columns[vectorized_data.dtypes == 'bool']:
    vectorized_data[col] = vectorized_data[col].astype(int)
```

```python
vectorized_data.head()
```

### Data Splitting

```python
from sklearn.model_selection import train_test_split

vec_train, vec_test = train_test_split(vectorized_data, random_state=42, train_size=0.8)

vec_train_y = vec_train['Clicked on Ad']
vec_train_x = vec_train.drop(columns=['Clicked on Ad'])
vec_test_y = vec_test['Clicked on Ad']
vec_test_x = vec_test.drop(columns=['Clicked on Ad'])
```

```python
print(vec_train_x.shape)
print(vec_train_y.shape)
print(vec_test_x.shape)
print(vec_test_y.shape)
```

### Vectorizing "Ad Topic Line" column

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
vector = tfidf.fit_transform(vec_train_x['Ad Topic Line'])
vector = pd.DataFrame(vector.toarray(), columns=tfidf.get_feature_names_out(), index=vec_train_x.index)
vec_train_x = pd.concat([vec_train_x, vector], axis=1)

vector2 = tfidf.transform(vec_test_x['Ad Topic Line'])
vector2 = pd.DataFrame(vector2.toarray(), columns=tfidf.get_feature_names_out(), index=vec_test_x.index)
vec_test_x = pd.concat([vec_test_x, vector2], axis=1)

vec_train_x = vec_train_x.drop(columns=['Ad Topic Line'])
vec_test_x = vec_test_x.drop(columns=['Ad Topic Line'])
```

```python
print(vec_train_x.shape)
print(vec_test_x.shape)
```

### Scaling Data

```python
from sklearn.preprocessing import StandardScaler

st_scaler = StandardScaler()
```

```python
vec_train_x = pd.DataFrame(st_scaler.fit_transform(vec_train_x), columns=vec_train_x.columns, index=vec_train_x.index)
vec_test_x = pd.DataFrame(st_scaler.transform(vec_test_x), columns=vec_test_x.columns, index=vec_test_x.index)
```

### PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=512)
```

```python
vec_train_x = pca.fit_transform(vec_train_x)
vec_test_x = pca.transform(vec_test_x)
```

#### In the preproccessing section we did the following steps respectively:
1. Omitting Extra columns: Before doing the visualization, we created two new columns, "date" and "time," by parsing the "TimeStamp" column. We then parsed these two new columns to create three additional columns: "hour," "day," and "month." After that, we omitted the "TimeStamp," "date," and "time" columns in order to train the model.
2. Drop duplicate rows: Duplicated rows in a dataset can cause the model to become overly confident in certain patterns or features, leading to biased predictions. When the model sees the same data points repeatedly, it may assign disproportionate importance to those features, even if they are not truly representative of the underlying patterns in the data. By removing the duplicated rows, we have ensured that the model is not unduly influenced by redundant information. This helps to create a more balanced and representative dataset, which in turn allows the model to learn the true relationships and patterns without being skewed by the presence of identical data points.
3. Encoding Categorical data: We had four categorical columns in our dataset: "City," "Gender," "Country," and "Ad Topic Line." For the first three columns - "City," "Gender," and "Country" - we used one-hot encoding to transform the categorical data into a numerical format that the model could understand. However, for the last column, "Ad Topic Line," since it consisted of text data, we decided to use a different approach. Instead of one-hot encoding, we used TF-IDF (Term Frequency-Inverse Document Frequency) to convert the text data into a numerical representation. TF-IDF is a common technique used in natural language processing to quantify the importance of words in a document or, in this case, the "Ad Topic Line" column.
$$ \text{TF} = \frac{\text{Number\: of\, times\: a\: word\: "X"\: appeares\: in\: a\: Document}}{\text{Number\: of\: words\: present\: in\: a\: Document}} $$
$$ \text{IDF} = \log(\frac{\text{Number\: of\, Document\: presents\: in\: a\: Corpus}}{\text{Number\: of\: Documents\: Where\: word\: "X"\: has\: appeared}}) $$
$$ \text{TF-IDF} = \text{TF} \times \text{IDF} $$
4. Data Scaling: In this part, we scaled the data using the Standard Scaler to prepare it for dimension reduction using Principal Component Analysis (PCA) and model training. The Standard Scaler is a common data preprocessing technique that transforms the features to have zero mean and unit variance. This is important because PCA and many machine learning models are sensitive to the scale of the features, and the Standard Scaler helps to ensure that all features are on a similar scale. By scaling the data using the Standard Scaler, we are normalizing the features so that they have a mean of 0 and a standard deviation of 1. This helps to ensure that features with larger numerical ranges do not dominate the analysis or model training process. It also helps to improve the stability and convergence of the PCA algorithm and the subsequent machine learning models.<br>
$$z = \frac{x - \mu}{\delta}$$
5. Dimension reduction: After encoding the categorical data and vectorizing the "Ad Topic Line" column using TF-IDF (Term Frequency-Inverse Document Frequency), the number of columns in the dataset has increased. However, the increase in the number of columns due to the TF-IDF transformation can lead to a high-dimensional feature space, which can be challenging for some machine learning models to handle effectively. High-dimensional data can suffer from the curse of dimensionality, where the model's performance may degrade as the number of features increases. To address this issue, we used Principal Component Analysis (PCA) to reduce the dimensionality of the dataset. PCA is a technique that identifies the principal components of the data, which are the directions of maximum variance. By selecting a subset of the principal components, we can reduce the dimensionality of the data while retaining most of the important information.

# Model Training

```python
from sklearn.metrics import f1_score, accuracy_score, confusion_matrix, roc_curve, ConfusionMatrixDisplay, RocCurveDisplay, auc
from sklearn.model_selection import GridSearchCV

def model_train_test_evaluating(x_train, y_train, x_test, y_test, model, param_grid, model_name):
    grid_search = GridSearchCV(model, param_grid=param_grid, cv=5, scoring='f1_macro')
    grid_search.fit(x_train, y_train)
    classifier = grid_search.best_estimator_

    y_train_pred = classifier.predict(x_train)
    y_test_pred = classifier.predict(x_test)

    train_accuracy = accuracy_score(y_train, y_train_pred)
    f1_train = f1_score(y_train, y_train_pred, average='macro')
    if not y_test.isnull().all():
        test_accuracy = accuracy_score(y_test, y_test_pred)
        f1_test = f1_score(y_test, y_test_pred, average='macro') 

    train_accuracy = accuracy_score(y_train, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    f1_train = f1_score(y_train, y_train_pred)
    f1_test = f1_score(y_test, y_test_pred) 

    print('Cross validation scores:')
    print(f"Best parameters: {grid_search.best_params_}")
    print(f"Best score: {grid_search.best_score_:.4f}")

    sns.set(style="white")
    print('train_data report:')
    cm = confusion_matrix(y_train, y_train_pred, labels=classifier.classes_)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classifier.classes_)
    fpr, tpr, thresholds = roc_curve(y_train, y_train_pred)
    roc_auc = auc(fpr, tpr)
    roc_display = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name=model_name)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    ax1.set_title('Confusion Matrix')
    disp.plot(ax=ax1)
    ax2.set_title('ROC Curve')
    roc_display.plot(ax=ax2)
    plt.subplots_adjust(wspace=0.5)
    plt.show()
    
    print('Acuuracy: ', end='')
    print(train_accuracy)
    print('F1_score: ', end='')
    print(f1_train)

    print('test_data report:')
    cm = confusion_matrix(y_test, y_test_pred, labels=classifier.classes_)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classifier.classes_)
    fpr, tpr, thresholds = roc_curve(y_test, y_test_pred)
    roc_auc = auc(fpr, tpr)
    roc_display = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name=model_name)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    ax1.set_title('Confusion Matrix')
    disp.plot(ax=ax1)
    ax2.set_title('ROC Curve')
    roc_display.plot(ax=ax2)
    plt.subplots_adjust(wspace=0.5)
    plt.show()

    print('Acuuracy: ', end='')
    print(test_accuracy)
    print('F1_score: ', end='')
    print(f1_test)
```

```python
from sklearn.linear_model import LogisticRegression

param_grid=[
    {'penalty': ['l1', 'l2']}
]

lr_classifier = LogisticRegression(penalty='l1', solver='saga')
model_train_test_evaluating(vec_train_x, vec_train_y, vec_test_x, vec_test_y, lr_classifier, param_grid, 'Logistic Regressor')
```

```python
from sklearn.svm import SVC

param_grid=[
    {'kernel': ['linear','poly', 'rbf']}
]

svm = SVC(kernel='linear')
model_train_test_evaluating(vec_train_x, vec_train_y, vec_test_x, vec_test_y, svm, param_grid, 'SVM')
```

```python
from sklearn.ensemble import RandomForestClassifier

param_grid=[
    {'n_estimators': [30, 40, 50]}
]

rf_classifier = RandomForestClassifier(random_state=42, criterion='gini')
model_train_test_evaluating(vec_train_x, vec_train_y, vec_test_x, vec_test_y, rf_classifier, param_grid, 'RandomForest')
```

#### In this part, as in the last part, we trained some classic classification models. To do this, we used the sklearn GridSearchCV class to perform k-fold cross-validation and to tune the model parameters. We set the number of folds, k, to 5, and for each model, we determined the parameter grid (paramgrid) according to the specific model we were using. The GridSearchCV class allows us to perform an exhaustive search over a specified parameter grid to find the best combination of parameters for the given model. By using k-fold cross-validation, we can ensure that the model is evaluated on different subsets of the data, providing a more robust estimate of the model's performance. The process of determining the parameter grid is crucial, as it defines the range of values to be explored for each hyperparameter of the model. The specific parameter grid will depend on the type of model being used, as different models have different sets of hyperparameters that can be tuned.

1. Logistic Regression:

#### It is obvious that our SVM model with an RBF (Radial Basis Function) kernel has performed better than the two other models, and it also seems that the RandomForest classifier has overfitted the data and could not perform well. The SVM model with an RBF kernel has demonstrated superior performance compared to the other models that were evaluated. This suggests that the non-linear decision boundary provided by the SVM model with the RBF kernel was more effective in capturing the underlying patterns in the data, leading to better classification accuracy. The RBF kernel is a popular choice for SVM models when the data is not linearly separable, as it can map the data into a higher-dimensional space where the classes become more linearly separable. On the other hand, the RandomForest classifier appears to have overfitted the training data, meaning that it has learned the training data too well and has failed to generalize to new, unseen data. This can happen when the model becomes too complex and captures noise or irrelevant features in the training data, rather than the true underlying patterns. As a result, the RandomForest classifier's performance on the test data was not as strong as the SVM model with the RBF kernel.
